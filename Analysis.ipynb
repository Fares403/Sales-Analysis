### Data Preprocessing , Cleaning And Transformation

```python
# change column Name branch_id to Branch_ID in orders
orders.rename(columns={'Branch ID': 'Branch_ID'}, inplace=True)
```

```python
# merge two dataframes
df_merged = pd.merge(orders, Branches, on='Branch_ID')
df_merged
```

```python
df=orders.merge(Branches, on='Branch_ID' , how='right')
df
```

```python
# add column Profit
df_merged['Profit'] = df_merged['Net Sales'] - df_merged['Cost']
```

```python
# check data
df_merged
```

```python
#make column branch_name as uppercase
df_merged['Branch_Name'] = df_merged['Branch_Name'].str.upper()
df_merged
```

### Answering Question With Are Data 

```python
# 1- total profit for each branch
df_merged.groupby('Branch_Name')['Profit'].sum()
```

```python
# 2- Number of Products
df_merged['Product-ID'].nunique()
```

```python
# 3-  Total Net Salary by Procut
df_merged.groupby('Product-ID')['Net Sales'].sum()
```

```python
# 4- top 10 products
df_merged["Product-ID"].value_counts().head(10)
```

```python
# 5- Top 2 Branches by Profit
df_merged.groupby('Branch_Name')['Profit'].sum().nlargest(2)
```

```python
# 6- Bottom 5 Products
df_merged['Product-ID'].value_counts().tail(5)
```

```python
# 7 - Number of Orers By Branch name 
df_merged.groupby('Branch_Name')['Invoice Number'].count()
```

```python
# 8 - # Number of Orders By Branch and Product
df_merged.groupby(['Branch_Name','Product-ID'])['Invoice Number'].count()
```

```python
# 9- Number of Invoices Only
df_merged['Invoice Number'].nunique()
```

```python
# 10 -Total Sales , AVG , MIN , MAX
total_sales = df_merged['Net Sales'].sum()
avg_sales = df_merged['Net Sales'].mean()
min_sales = df_merged['Net Sales'].min()
max_sales = df_merged['Net Sales'].max()
print(total_sales,avg_sales,min_sales,max_sales , sep = "\n")
```
